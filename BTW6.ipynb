```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
import numpy as np


url ='all_stocks_5yr.csv'
data = pd.read_csv(url)
data.head()

data.hist()
plt.show()
sns.pairplot(data)
```

```python
correlation = data .corr()
print(correlation)
sns.heatmap(correlation, annot = True)
```

```python
# bậc 1
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(pd.DataFrame(data['volume']), data['close'])
print ('Gia tri intercept: ',model.intercept_)
print('Gia tri coffient: ', model.coef_)
print('Do chinh xac: ', model.score(X= pd.DataFrame(data['volume']),y = data['close']))
```

```python
data.plot(kind = 'scatter', x = 'volume', y = 'close', figsize = (9,9), color = 'black' )
y_pred = model.predict(X=pd.DataFrame(data['volume']))
plt.plot(data['volume'],y_pred,color='blue')
```

```python
#bậc 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['volume'],data['volume']**2]).T
model.fit(X,y = data['close'])
print('Intercept la: ', model.intercept_)
print('Coefficient la: ', model.coef_)
print('Do chinh xac:', model.score(X,y = data['close']))
```

```python
range = np.arange(1,6,0.05)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter', x = 'volume' , y = 'close', figsize = (9,9), color ='black')
plt.plot(range,y_pred , color ='blue')
```
